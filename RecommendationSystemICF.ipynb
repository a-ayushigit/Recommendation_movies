```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('/content/movie_new_df.csv')
```

```python
df.head()
```

```python
df.drop('Unnamed: 0' , axis=1 , inplace=True)
```

```python
#The movie Ids and user Ids are not continuous
```

```python
pt = df.pivot_table(index='title' , columns='userId' , values='rating')
```

```python
pt
```

```python
pt_centered = pt.sub(pt.mean(axis=0) , axis=1)
```

```python
pt_centered
```

```python
pt_centered.fillna(0 , inplace=True)
```

```python
similarity_scores = cosine_similarity(pt_centered)
```

```python
similarity_scores.shape
```

```python
df_cleaned = df.drop_duplicates('title')
```

```python
def recommend(movie):
  index = np.where(pt.index == movie)[0][0]
  similar_movies = sorted(list(enumerate(similarity_scores[index])), key=lambda x: x[1], reverse=True)[1:6]

  data = []
  for movie_index, _ in similar_movies:
    data.append(df_cleaned.iloc[movie_index]['title'])

  return data
```

```python
recommend('Kung Fu Panda (2008)')
```
